```python
import os
os.chdir(filePath)
arr_txt = [x for x in os.listdir() if x.endswith(".csv")and x !='Covid19_confirmed_cases.csv']
print(arr_txt)
```

```python
import pandas as pd

all_DF=[]
for file in arr_txt:
  #url="https://github.com/CSSEGISandData/COVID-19/blob/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv"
  print(filePath + '/' + file)
  tmp_DF = pd.read_csv(filePath + '/' + file)
  try:
    all_DF.append(tmp_DF[tmp_DF['Country_Region']=='India'])
  except:
    all_DF.append(tmp_DF[tmp_DF['Country/Region']=='India'])

res = pd.concat(all_DF)
#res=merged_df.loc[merged_df['Country_Region']=='India',['Last_Update','Confirmed','Deaths','Recovered']]
res.to_csv(filePath + '/ConsolidatedDataIndia.csv')
print('Completed')
```
