```python
#Activity 5 - Image detection from Video images
#func1_getGreyImg(img)
#func2_faceDetect(img_dir)
#func3_faceRecog(faces,faceid)
#func4_drawRect(test_img,face)
#func5_updLabel(test_img,text,x,y)

facerecognition = cv2.face.LBPHFaceRecognizer_create()
facerecognition.read(r"C:\Users\DELL\Documents\PAI-0221A\project\face_training.yml")

name = {1:"Ben Affleck",2:"Elton John",3:"Brad Pitt",4:"Mindy Kaling",5:"Rajesh"}

Cap = cv2.VideoCapture(0)

stopCam = False
while True:
    ret,Img = Cap.read()
    
    FacesDetected, GreyImg = f1.func1_getGreyImg(Img)
    
    for face in FacesDetected:
        (x,y,w,h) = face
    
        GrayImage = GreyImg[y:y+w,x:x+h]
        label,confidence = facerecognition.predict(GrayImage)
    
        print("Confidence", confidence)
        print("Label",label)
    
        f4.func4_drawRect(Img,face)
        predicted_name = name[label]
    
        #testing
        print("Predicted Name:", predicted_name)
        #testing 
    
        #if confidence level less than 40% continue
        if confidence < 20:
            continue
        else:
            f5.func5_updLabel(Img,predicted_name,x,y)
            resizedimage = cv2.resize(Img,(600,500))
        
            #Testing outputs
            while True:
                cv2.imshow("Cropped Final Image", resizedimage)
                key=cv2.waitKey(1)
                if key == 13:
                    break
                elif key == 27:
                    stopCam=True
                    break
        
    if stopCam == True:
        break
Cap.release()
cv2.destroyAllWindows()
        

```
